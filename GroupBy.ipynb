### group by 메서드
- 분리 - 적용 - 경합
- 특정 컬럼의 값 혹은 인덱스를 group 화
- groupby 만 수행시 분리만 수행, 적용함수 따로 전달

```python
emp = pd.read_csv('emp.csv', engine='python')
emp
```

```python
emp.groupby('DEPTNO').sum()   # 연산을 지정하지않으면 분리데이터 셋 저장만 한 상태가 된다 
                              # groupby 컬럼은 인덱스로 출력된다.
```

```python
emp.groupby('DEPTNO')['SAL'].sum()  # deptno 별로 sal 의 총합  Series 형태로 출력 
```

```python
emp.groupby('DEPTNO')[['SAL']].sum()  # DataFrame 형태로 출력
```

```python
emp.groupby(['DEPTNO','JOB'])[['SAL']].sum()  # 여러개 컬럼 연산시 속도가 느릴수 있음
```

```python
emp.groupby(level=0).mean()  # index의 특정 level groupby 가능
```

```python
emp.groupby('DEPTNO')[['SAL']].sum().reset_index()
```

```python
emp.groupby('DEPTNO')[['SAL','COMM']].agg(['sum','mean'])  # 한가지 이상의 연산 전달 가능 
```

```python
emp.groupby('DEPTNO')[['SAL','COMM']].agg({'SAL':'sum','COMM':'mean'})  # 딕셔너리형태로 바꾸면서 각각 전달 가능
```
