## Multi-index 2 

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from pandas import Series
```

```python
# 1. multi_index_ex1.csv 파일을 일고
df1 = pd.read_csv('multi_index_ex1.csv', engine='python')
df1
```

```python
# 멀티컬럼 설정
df1.columns.map(lambda x : x[:2])
df1.iloc[0,:]
df1.columns = [df1.columns.map(lambda x : x[:2]),df1.iloc[0,:]]
df1
```

```python
# 첫번째 행 제외
df1 = df1.iloc[1:,]
```

```python
# 멀티 인덱스 설정
df1.iloc[:,0]
df1.iloc[:,1]
df1.index=[df1.iloc[:,0],df1.iloc[:,1]]
df1.index
df1 = df1.iloc[:,2:]
df1
```

```python
# 인덱스 이름 설정
df1.index.names=['구분','지점']
df1.columns.names=['지역','지점']
df1
```

```python
# 2) 컴퓨터의 서울지역 판매량 출력
df1.loc['컴퓨터','서울']
```

```python
# 3) 서울지역의 컴퓨터이 각 세무항목별 판매량의 합계 출력
df1 = df1.astype('int')
df1.loc['컴퓨터','서울'].sum(1)
```

```python
# 4) 각 지역의 A 지점의 TV 판매량 출력
df1.xs('A',axis=1,level=1).xs('TV',axis=0,level=1)
```

```python
# 5) 각 지역의 C 지점의 모바일의 각 세무항목별 한매량 평균 출력
df1.xs('C',axis=1,level=1).loc['모바일',:].sum(1)
```

```python
# 6) 서울 지역의 A 지점의 노트북 판매량 출력 
df1.xs('노트북',axis=0,level=1).loc[:,'서울']
```

```python
# 지역별 판매량 총 합
df1.sum(axis=1, level=0)
```

```python
# 구분(컴퓨터, 가전, 모바일)별, 지역별 판매량 총 합
df1.sum(axis=0,level=0).sum(axis=1,level=0)
```
