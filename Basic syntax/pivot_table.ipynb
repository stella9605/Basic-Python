## cross-table
- 행별, 열별 연산 용이
- join 불가
- group by 연산 불가
- 파이썬 시각화 시 주로 사용
- multi-index 를 갖는 구조를 unstack처리하여 얻거나 pivot 을 통해 얻을 수 있음

## pivot
- 각 컬럼의 값을 교차테이블 구성요소로 전달, 교차테이블 완성
- index, columns, values 컬럼을 각각 전달
- 요약기능 없음

## pivot_table
- pivot 기능과 유사, 더 많은 옵션 사용 가능

## - index, columnms, values 컬럼을 각각 전달
- aggfunc 옵션 사용하여 요약 기능 전달 가능(기본은 평균)
- fill_value 옵션 사용하여 NA값 대체 가능

```python
delivery = pd.read_csv('delivery.csv', engine = 'python')
```

```python
# 1) 각 업종별 통화건수가 많은 순서대로 시군구의 순위를 출력
delivery.pivot_table(index = '시간대',
                     columns = '업종',
                     values = '통화건수',
                     aggfunc = 'sum').rank(axis=0, ascending=False, method='min')
```

```python
 # 다른방법 

df2 = delivery.set_index(['시간대','업종'])['통화건수'].sum(level=[0,1]).unstack()
df2
```

```python
df2.rank(axis=0, ascending=False, method = 'min')
```

```python
# 시간대 별 배달 콜수가 가장 많은 업종 1개 출력
df3 = delivery.pivot_table(index = '시간대',
                    columns = '업종',
                    values='통화건수',
                    aggfunc='sum')
df3
```
